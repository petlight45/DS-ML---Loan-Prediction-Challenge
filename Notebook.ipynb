```python
default_params = {
#     "limit":1,
    'access_key':'e4f1106101f23e261697705f9a0fdec5'
}
reverse_geocoding_endpoint = 'http://api.positionstack.com/v1/reverse'
def reverse_geo(*lat_long):
    query_param = {
    'query':','.join(lat_long)
    }
    query_param.update(default_params)
    response = requests.get(reverse_geocoding_endpoint, params=query_param)
    return response.json()
    
```

```python
reverse_geo_data = demo[['customerid']]
```

```python
len_data = reverse_geo_data.shape[0]
c_length = 1
```

```python
def test(x):
    global c_length
    lat = str(demo.set_index('customerid').loc[x['customerid']]['latitude_gps'])
    long = str(demo.set_index('customerid').loc[x['customerid']]['longitude_gps'])
    trials_left = 10
    while True:
        try:
            trials_left -= 1
            reverse_data = reverse_geo(lat,long)['data'][0]
            x['Country'] = reverse_data['country']
            x['State'] = reverse_data['region']
            x['LGA'] = reverse_data['county']
            break
        except:
            if trials_left:
                continue
            else:
                x['Country'] = np.nan
                x['State'] = np.nan
                x['LGA'] = np.nan
                break
    print(f"{c_length}/{len_data}")
    c_length += 1
    return x
```

```python
reverse_df = reverse_geo_data.apply(test, axis=1)
```

```python
geo_test = pd.read_csv('demo_revgeo_test.csv')
geo_train = pd.read_csv('demo_revgeo_train.csv')
```

```python
geo_train.drop('Unnamed: 0', axis=1,inplace=True)
geo_test.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
def fill_na_(data):
    data = data.copy()
    data['Country'] = data['Country'].fillna("Unknown Country")
    data['State'] = data['State'].fillna("Unknown State")    
    data['LGA'] = data['LGA'].fillna("Unknown LGA")       
    return data
```

```python
geo_train = fill_na_(geo_train)
geo_test = fill_na_(geo_test)
```
